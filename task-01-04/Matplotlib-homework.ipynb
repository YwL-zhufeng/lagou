```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取数据
data_man = pd.read_excel('./体测分数_男生.xls')
data_woman = pd.read_excel('./体测分数_女生.xls')
display(data_man.head(),data_woman.head())
```

```python
#对男1000米跑、男引体分数进行等宽分箱操作，分成3份
labels = ['低','中','高'] # 三组标签
dm_runk = pd.cut(data_man['男1000米跑分数'],3,labels=labels)
dm_up = pd.cut(data_man['男引体分数'],3,labels=labels)
display(dm_runk.head(),dm_up.head())
```

```python
# 绘制男1000米跑分数饼图

plt.figure(figsize=(9,9))
_ = plt.pie([dm_runk[dm_runk == '高'].count(),dm_runk[dm_runk == '中'].count(),dm_runk[dm_runk == '低'].count()],
        labels=labels,textprops={'family':'KaiTi','fontsize':18},
        autopct='%0.2f%%',# 显示概率 
        explode = [0,0,0.2], # 突出显示成绩为高
        shadow=True, # 加阴影
       )
plt.title(label='男1000米跑分数占比',
          fontdict={'family':'KaiTi','fontsize':18},
          pad = 30)
```

```python
# 绘制男引体分数饼图

plt.figure(figsize=(9,9))
_ = plt.pie([dm_up[dm_up == '高'].count(),dm_up[dm_runk == '中'].count(),dm_up[dm_up == '低'].count()],
        labels=labels,textprops={'family':'KaiTi','fontsize':18},
        autopct='%0.2f%%',# 显示概率 
        explode = [0,0,0.2], # 突出显示成绩为高
        shadow=True, # 加阴影
       )
plt.title(label='男引体分数占比',
          fontdict={'family':'KaiTi','fontsize':18},
          pad = 30)
```

```python
#对女800米跑进行直方图绘制统计各分数段人数，分成4份
labels = ['差','中','良','优']
dw_r = pd.cut(data_woman['女800米跑分数'],4,labels=labels)
dw_r_count = dw_r.value_counts()

plt.rcParams['font.family'] = 'KaiTi' # 全局设置字体
plt.rcParams['font.size'] = 18 # 全局设置字体大小

#画图
plt.figure(figsize=(9,6))

plt.bar(x=dw_r_count.index,
       height=dw_r_count,
       width = 0.3,
       )
plt.grid(axis='y')

# 添加注释 显示每个数据
for id in dw_r_count.index:
    plt.text(id, dw_r_count[id]+ 0.5,dw_r_count[id], ha='center', va='bottom', fontsize=18)

_ = plt.title(label='女800米跑各分段人数',
          fontdict={'family':'KaiTi','fontsize':18},
          pad = 30)
```

```python
#对女跳远进行直方图绘制统计各分数段人数，分成4份
labels = ['差','中','良','优']
dw_r = pd.cut(data_woman['女跳远分数'],4,labels=labels)
dw_r_count = dw_r.value_counts()

plt.rcParams['font.family'] = 'KaiTi' # 全局设置字体
plt.rcParams['font.size'] = 18 # 全局设置字体大小

#画图
plt.figure(figsize=(9,6))

plt.bar(x=dw_r_count.index,
       height=dw_r_count,
       width = 0.3,
       )
plt.grid(axis='y')

# 添加注释 显示每个数据
for id in dw_r_count.index:
    plt.text(id, dw_r_count[id]+ 0.5,dw_r_count[id], ha='center', va='bottom', fontsize=18)
    
_ = plt.title(label='女跳远各分段人数',
          fontdict={'family':'KaiTi','fontsize':18},
          pad = 30)
```

```python
#使用嵌套饼图对比男女生体重指数BMI进行比例统计，分为正常、低体重、超重、肥胖
labels = ['低体重','正常','超重','肥胖']
dm_bmi = pd.cut(data_man['BMI'],bins=[0,16.4,23.2,26.3,100],labels=labels)
dw_bmi = pd.cut(data_woman['BMI'],bins=[0,16.4,22.7,25.2,100],labels=labels)
dm_bmi_count = dm_bmi.value_counts()
dw_bmi_count = dw_bmi.value_counts()

plt.figure(figsize=(9,9))

cs = [ 'blue', 'green', 'orange', 'red']

plt.pie(dm_bmi_count,
        radius=1, # 半径
        autopct='%0.2f%%', # 显示百分比
        pctdistance=0.85, # 调整概率位置
        #labels = labels,
        colors = cs,
        wedgeprops={'linewidth':5,# 间隔宽度
                   'width':0.3,# 饼图宽度
                   'edgecolor':'white'},# 间隔颜色
       textprops = {'family':'KaiTi','fontsize':18}, 
       )

_ = plt.pie(dw_bmi_count,
        radius=0.7,
        autopct='%0.2f%%',
        pctdistance=0.55,
        colors = cs,
        wedgeprops={'linewidth':5,# 间隔宽度
                   'width':0.7,# 饼图宽度
                   'edgecolor':'white'},# 间隔颜色
       )

plt.legend(labels=labels,
            loc='best',
           fontsize = 14)

_ = plt.title(label='男女生体重指数BMI',
              fontdict={'family':'KaiTi','fontsize':18},
              pad = 30)
```
