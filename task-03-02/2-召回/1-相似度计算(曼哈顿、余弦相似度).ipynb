## 相似度计算

```python
users = {
    "Angelica":{"星际穿越":3.5,
                "放牛班的春天":2.0,
                "肖申克的救赎":4.5,
                "盗梦空间":5.0,
                "阿甘正传":1.5,
                "乱世佳人":2.5,
                "傲慢与偏见":2.0},
    "Bill":{"星际穿越":2.0,
            "放牛班的春天":3.5,
            "肖申克的救赎":4.5,
            "盗梦空间":5.0,
            "阿甘正传":1.5,
            "乱世佳人":2.5,
            "傲慢与偏见":2.0},
    "Chan":{"星际穿越":2.0,
            "放牛班的春天":3.5,
            "肖申克的救赎":3.5,
            "盗梦空间":5.0,
            "阿甘正传":2.5,
            "乱世佳人":2.5,
            "傲慢与偏见":2.0},
}
```

### 曼哈顿距离计算相似度

```python
# 曼哈顿距离计算相似度
def man_dis(user1,user2):
    
    distance = 0
    flag = False
    
    # 把每个用户与其他用户共有的电影找到
    for m in users[user1]:
        if m in users[user2]: # 判断电影是否在用户2中，表示有相似，然后在进行相似度计算
            distance += abs(users[user1][m]-users[user2][m])  # 累加求和
            flag = True
    # 判断是否完全不相似，即用户1里的电源都不存在在用户2中，此时相似度不能为0
    if flag == True:
        return distance
    else:
        return -1
```

```python
man_dis('Bill','Chan')
```

### 余弦相似度

```python
import numpy as np
```

```python
# 余弦相似度
def cos_dis(user1,user2):
    
    rating1 = users[user1]
    rating2 = users[user2]
    
    # 取出用户1存在在用户2的电影
    common_movies = [x for x in rating1 if x in rating2]
    
    # 若无共同电影
    if len(common_movies) == 0:
        return -1
    
    # 构建用户评分，使用列表生成式
    fenzi = sum([rating1[x] * rating2[x] for x in common_movies]) # 电影得分相乘累和 分子
    
    # 计算模长 即分母
    len_1 = np.sqrt(sum([rating1[x] **2 for x in common_movies]))
    len_2 = np.sqrt(sum([rating2[x] **2 for x in common_movies]))
    
    # 避免分母为零，即两个用户对两个电影打零分
    if len_1 + len_2 == 0:
        return -1
    
    return fenzi / (len_1 * len_2)
```

```python
cos_dis('Bill','Chan')
```
