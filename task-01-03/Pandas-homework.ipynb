```python
# 构造跑步打分函数
def grade_run(mark,performance):
    score = [] # 保存成绩
    for perform in performance:
        if perform <= mark['成绩'][0]: # 大于最大值 直接满分
            score.append(mark['分数'][0])
            continue
        if perform > mark['成绩'].iloc[-1]: # 小于最小值 直接0分
            score.append(0)
            continue
        for j in range(len(mark)-1):  
            if mark['成绩'][j] < perform <= mark['成绩'][j+1]:# ( , ] 区间内成绩
                score.append(mark['分数'][j+1])
                break
    return score
```

```python
# 计算男1000米跑成绩
score = grade_run(data_mark.loc[:,'男1000米跑'],data_man['男1000米跑'])
data_man['男1000米跑成绩'] = score
data_man
```

```python
# 计算男50米跑成绩
score = grade_run(data_mark.loc[:,'男50米跑'],data_man['男50米跑'])
data_man['男50米跑成绩'] = score
data_man
```
